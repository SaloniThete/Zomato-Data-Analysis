```python
country_name = df_final.Country.value_counts().index
country_name
```

```python
country_value = df_final.Country.value_counts().values
country_value
```

**Top 3 countries that use zomato using pie chart**

```python
plt.pie(country_value[:3], labels = country_name[:3],autopct = '%1.2f%%')
plt.show()
```

**Observation**

*Maximum records or transactions of zomato are from India after USA and UK*

```python
ratings = df_final.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:"Rating Count"})
ratings
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(8,4)
sns.barplot(x="Aggregate rating",y="Rating Count", data=ratings)
```

```python
import matplotlib
matplotlib.rcParams['figure.figsize']=(8,4)
sns.barplot(x="Aggregate rating",y="Rating Count",hue='Rating color',
           data=ratings,palette=['blue','red','orange','yellow','green','green'])
```

**Count Plot**

```python
sns.countplot(x = "Rating color", data = ratings, palette=['blue','red','orange','yellow','green','green'])
```
